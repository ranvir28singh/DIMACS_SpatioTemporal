```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
words = open('Files/emotions2.txt')
words = words.readlines()

words = [i.split(':') for i in words]
#test = [i.replace(' \n','') for i in test]
#test = [i.replace(':',',') for i in test]
#test = [i.replace(' ',',') for i in test]
#test = [i.split(',') for i in test]

for i in words:
    i[0] = i[0].replace(' ','|')
    i[1] = i[1].replace(' \n','')
    i[1] = i[1].replace(' ',',')
    
data = []

for i in words:
    d = i[0] + ',' + i[1]
    d = d.split(',')
    data.append(d)
    
df_emotions = pd.DataFrame(data)

df_emotions.rename(columns={0:'word'},inplace=True)
columns = ['attrib_'+str(i) for i in range(1,301)]
df_emotions.columns.values[1:301] = columns

df_emotions['word'] = df_emotions['word'].map(lambda x : x.replace('|',' '))

word = df_emotions['word']
df_emotions.drop(labels=['word'], axis=1,inplace = True)
df_emotions.insert(300, 'word', word)
df_emotions
```

```python
common_words = open('Files/common.txt')
common_words = common_words.readlines()

common_words = [i.split(':') for i in common_words]

for i in common_words:
    i[0] = i[0].lower()
    i[0] = i[0].replace('_',' ')
    i[0] = i[0].replace(' ','|')
    i[0] = i[0].replace('.','')
    i[0] = i[0].replace("'",'')
    i[1] = i[1].replace(' \n','')
    i[1] = i[1].replace(' ',',')
    

data = []

for i in common_words:
    if i[0].isalnum():
            d = i[0] + ',' + i[1]
            d = d.split(',')
            data.append(d)
    
df_common = pd.DataFrame(data)

df_common.rename(columns={0:'word'},inplace=True)
columns = ['attrib_'+str(i) for i in range(1,301)]
df_common.columns.values[1:301] = columns

df_common['word'] = df_common['word'].map(lambda x : x.replace('|',' '))

word = df_common['word']
df_common.drop(labels=['word'], axis=1,inplace = True)
df_common.insert(300, 'word', word)
df_common.drop_duplicates(subset='word',inplace=True)
df_common.head(5)
```

```python
df_common.to_csv('common_processed.csv',float_format='%.5f',index=False)
```

```python
## Import cluster results from WEKA
import scipy.io.arff as arff

cluster_arff = arff.loadarff('word-cluster-10.arff')

df_clusters = pd.DataFrame.from_records(cluster_arff[0])
df_clusters.Cluster = df_clusters.Cluster.map(lambda x: x.decode('UTF-8'))
df_clusters.word = df_clusters.word.map(lambda x: x.decode('UTF-8'))
```
