```python
# Get unique entrances just to pick one and see if we can do something with it
combined_df.entrance.unique()
```

```python
# For now, group the values by timestamp and get the means of X Y Z etc.
# We could further add more granularity by also adding agent type or disabled/luggage in the 'by' argument
# Let's just look at one entrance and try to see what 'average' path those agents take over time 
```

```python
test_df = combined_df[combined_df['entrance'] == 'entry_exit_591'].groupby(by='sec')[['X','Y','Z','velocity','lookUp_X','lookUp_Y']].mean()
```
